## Central limit Theorm

As the sample size increases, the distribution of sample means of a random variable will tend to be normal or Gaussian in shape, regardless of the underlying distribution of the original variable.

If we take a large enough sample size from any population, the sample mean will be normally distributed regardless of the shape of the original population distribution

 

(img source - https://www.youtube.com/watch?v=PUBZC2MJ50Y)

```python
import numpy as np
import matplotlib.pyplot as plt

sample_size = 50
num_samples = 10000

# Generate the samples from a non-normal distribution (exponential)
samples = np.random.exponential(scale=2, size=(num_samples, sample_size))

sample_means = np.mean(samples, axis=1)

# Plot the histogram of the sample means
plt.hist(sample_means, bins=30)
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Central Limit Theorem Demonstration')
plt.show()
```
